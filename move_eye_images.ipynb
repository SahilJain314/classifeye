```python
import os

image_path = "D:deep/mrlEyes/"

files = os.listdir(image_path)
open = 0
closed = 0


for f in files:
    if f.endswith('png'):
        if f[16] == '1':
            os.rename(image_path+f,image_path+'open/'+f)
        else:
            os.rename(image_path+f,image_path+'closed/'+f)
            
print(open)
print(closed)
```
